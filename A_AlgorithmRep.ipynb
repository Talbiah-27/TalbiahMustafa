### Weighted Graph and Heuristic Values

This section defines a sample weighted graph and associated heuristic values. A weighted graph has numerical costs (weights) associated with its edges, representing the 'cost' of traversing that edge. Heuristic values are estimates of the cost from a given node to the goal node, often used in search algorithms like A*.

```python
# Define a weighted graph using an adjacency list
# Each key is a node, and its value is a dictionary of neighbors with their edge weights.
weighted_graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'A': 1, 'D': 5, 'E': 2},
    'C': {'A': 4, 'F': 2},
    'D': {'B': 5, 'G': 3},
    'E': {'B': 2, 'G': 6},
    'F': {'C': 2, 'G': 1},
    'G': {}
}

# Define heuristic values for each node to a hypothetical goal node (e.g., 'G')
# These values are estimates of the cost from each node to the goal.
heuristic_values = {
    'A': 7,
    'B': 6,
    'C': 3,
    'D': 2,
    'E': 3,
    'F': 1,
    'G': 0  # Heuristic for the goal node is typically 0
}

print("Weighted Graph:")
for node, neighbors in weighted_graph.items():
    print(f"  {node}: {neighbors}")

print("\nHeuristic Values (to goal 'G'):")
for node, heuristic in heuristic_values.items():
    print(f"  {node}: {heuristic}")
```

### A* Search Algorithm

The A* (A-star) search algorithm is a popular pathfinding and graph traversal algorithm, widely used in artificial intelligence and robotics. It's an informed search algorithm, meaning it uses heuristic information to guide its search.

#### How A* Works:

A* finds the shortest path between a starting node and a goal node in a graph. It does this by evaluating each node using a cost function `f(n) = g(n) + h(n)`:

*   **`g(n)`**: The cost of the path from the starting node to node `n`.
*   **`h(n)`**: The heuristic estimate of the cost from node `n` to the goal node. This is the 'informed' part of the search, guiding the algorithm towards the goal.

A* maintains two sets of nodes:

1.  **Open Set**: Nodes that have been visited but not yet evaluated. These are candidates for expansion.
2.  **Closed Set**: Nodes that have already been evaluated.

It prioritizes nodes in the open set with the lowest `f(n)` value, ensuring it explores the most promising paths first. If the heuristic `h(n)` is *admissible* (never overestimates the true cost to the goal) and *consistent* (satisfies the triangle inequality), A* is guaranteed to find the optimal (shortest) path.

```python
import heapq

def a_star_search(graph, heuristics, start, goal):
    # The open set contains tuples of (f_score, node, g_score, path)
    # f_score: g_score + h_score (total estimated cost)
    # g_score: actual cost from start to current node
    # path: list of nodes from start to current node
    open_set = []
    heapq.heappush(open_set, (heuristics[start], start, 0, [start]))

    # keep track of visited nodes and their minimum g_score found so far
    g_scores = {node: float('inf') for node in graph}
    g_scores[start] = 0

    closed_set = set()

    while open_set:
        f_score, current_node, g_score, path = heapq.heappop(open_set)

        if current_node == goal:
            return path, g_score # Return the path and its total cost

        if current_node in closed_set:
            continue
        closed_set.add(current_node)

        for neighbor, weight in graph[current_node].items():
            tentative_g_score = g_score + weight

            if tentative_g_score < g_scores[neighbor]:
                g_scores[neighbor] = tentative_g_score
                h_score = heuristics[neighbor]
                f_score = tentative_g_score + h_score
                new_path = path + [neighbor]
                heapq.heappush(open_set, (f_score, neighbor, tentative_g_score, new_path))

    return None, float('inf') # No path found
```

### Applying A* to Our Graph

Now let's use the A* algorithm to find the shortest path from a starting node (e.g., 'A') to the goal node ('G') in our defined `weighted_graph`.

```python
start_node = 'A'
goal_node = 'G'

path, cost = a_star_search(weighted_graph, heuristic_values, start_node, goal_node)

if path:
    print(f"Shortest path from {start_node} to {goal_node}: {path}")
    print(f"Total cost: {cost}")
else:
    print(f"No path found from {start_node} to {goal_node}")
```
