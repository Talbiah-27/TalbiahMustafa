### Easy Example: Finding the peak in a simple list

Let's apply the `simple_hill_climbing` function to find the maximum value within a small, predefined list of numbers. The 'solution' in this case will be the index within the list.

```python
# A very simple list to search for a peak
data_list = [1, 2, 5, 9, 7, 3, 0]
print(f"Data list: {data_list}")

# Objective function: return the value at the given index
def objective_list(index):
    return data_list[index]

# Get neighbors (adjacent indices)
def get_list_neighbors(index):
    neighbors = []
    if index > 0:
        neighbors.append(index - 1)
    if index < len(data_list) - 1:
        neighbors.append(index + 1)
    return neighbors

# Initial solution: a random index within the list
initial_index_list = random.randint(0, len(data_list) - 1)
print(f"Starting hill climbing on the list from index: {initial_index_list}")

# Run the hill climbing algorithm
best_index_found, best_value_found = simple_hill_climbing(
    objective_function=objective_list,
    initial_solution=initial_index_list,
    get_neighbors=get_list_neighbors
)

print(f"\nBest index found: {best_index_found}")
print(f"Maximum value found: {best_value_found} (which is data_list[{best_index_found}])")
```

## Simple Hill Climbing Algorithm

Simple Hill Climbing is a local search algorithm that iteratively moves in the direction of increasing objective function value (for maximization problems) or decreasing objective function value (for minimization problems). It terminates when it reaches a "peak" or "valley" where no neighboring state has a better objective value.

**Key Characteristics:**
*   **Greedy Approach:** It always chooses the best immediate neighbor.
*   **Local Optima Trap:** It can get stuck in local optima and may not find the global optimum.
*   **No Backtracking:** It does not remember previous states, making it susceptible to repeated states if not handled.
*   **Simplicity:** It's easy to understand and implement.

```python
import random

def simple_hill_climbing(objective_function, initial_solution, get_neighbors, max_iterations=1000):
    """
    Implements the Simple Hill Climbing algorithm.

    Args:
        objective_function (callable): The function to maximize. Takes a solution as input.
        initial_solution (list or tuple): The starting point for the search.
        get_neighbors (callable): A function that takes a solution and returns a list of its neighbors.
        max_iterations (int): Maximum number of iterations to prevent infinite loops.

    Returns:
        list or tuple: The best solution found.
    """
    current_solution = initial_solution
    current_value = objective_function(current_solution)

    for _ in range(max_iterations):
        neighbors = get_neighbors(current_solution)
        if not neighbors:
            break # No neighbors, cannot move further

        best_neighbor = None
        best_neighbor_value = current_value

        for neighbor in neighbors:
            neighbor_value = objective_function(neighbor)
            if neighbor_value > best_neighbor_value: # For maximization
                best_neighbor_value = neighbor_value
                best_neighbor = neighbor

        if best_neighbor is None or best_neighbor_value <= current_value:
            # No better neighbor found, we are at a local optimum
            break
        else:
            current_solution = best_neighbor
            current_value = best_neighbor_value

    return current_solution, current_value


# --- Example Usage ---

# 1. Define an Objective Function (to maximize)
# Let's try to maximize f(x) = -(x-5)^2 + 10. The peak is at x=5, f(5)=10.
def objective_function(x):
    return -(x - 5)**2 + 10

# 2. Define a function to get neighbors
# For an integer x, neighbors are x-1 and x+1
def get_int_neighbors(x):
    return [x - 1, x + 1]

# 3. Set an initial solution
initial_point = random.randint(0, 10) # Start from a random point between 0 and 10
print(f"Starting hill climbing from: x = {initial_point}")

# 4. Run the hill climbing algorithm
best_solution, best_value = simple_hill_climbing(
    objective_function=objective_function,
    initial_solution=initial_point,
    get_neighbors=get_int_neighbors
)

print(f"\nBest solution found: x = {best_solution}")
print(f"Maximum value at this solution: f(x) = {best_value}")

# --- Another Example: Maximizing a 2D function ---

def objective_function_2d(coords):
    x, y = coords
    return -(x-2)**2 - (y-3)**2 + 20 # Peak at (2,3), f(2,3)=20

def get_2d_neighbors(coords):
    x, y = coords
    neighbors = [
        (x + 1, y), (x - 1, y), (x, y + 1), (x, y - 1), # Cardinal directions
        (x + 1, y + 1), (x - 1, y - 1), (x + 1, y - 1), (x - 1, y + 1) # Diagonal directions
    ]
    # Filter to keep within some bounds, if necessary (not strictly needed for this example, but good practice)
    return [(nx, ny) for nx, ny in neighbors if -10 <= nx <= 10 and -10 <= ny <= 10]

initial_point_2d = (random.randint(-5, 5), random.randint(-5, 5))
print(f"\nStarting 2D hill climbing from: (x, y) = {initial_point_2d}")

best_solution_2d, best_value_2d = simple_hill_climbing(
    objective_function=objective_function_2d,
    initial_solution=initial_point_2d,
    get_neighbors=get_2d_neighbors
)

print(f"\nBest 2D solution found: (x, y) = {best_solution_2d}")
print(f"Maximum value at this solution: f(x, y) = {best_value_2d}")
```
